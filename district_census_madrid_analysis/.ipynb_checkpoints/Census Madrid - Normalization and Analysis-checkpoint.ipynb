# Normalize and Scale data per districts and segments

Normalise data total men or spanish men (no correlation between these variables and districts) with the total men or total spanish men in each neighbourhood and plot heat map for each neighbourhood year by year.

```python
from config import *
import itertools
import scipy.stats as st
import pandas as pd
import numpy as np
from os import listdir
from os.path import isfile, join
import matplotlib.pyplot as plt  
import matplotlib.gridspec as gridspec
import seaborn as sns
from sklearn.preprocessing import StandardScaler
%matplotlib inline
```

### 1.Normalize and Scale data

```python
data = pd.read_csv('complete_clean_census_madrid.csv')
data = data.rename(columns={'Unnamed: 0':'age_range'})
```

```python
def normalize_inhabitants_per_year(inhabitants_per_year, data, columns_to_normalize):
    
    normalized_data = []
    for year in inhabitants_per_year.year.unique().tolist():
        total_inh = inhabitants_per_year[inhabitants_per_year['year'] == year]['total'].values[0]
        segment_year = data[data['year'] == year]
        for column in columns_to_normalize:
            segment_year[column] = segment_year[column]/total_inh
        normalized_data.append(segment_year)  
    
    return pd.concat(normalized_data)
```

#### 1.1. Normalize each segment by total number of inhabitants per year

```python
inhabitants_per_year = data[data['age_range'] == 'Total'].loc[:,['year','total']].groupby(['year']).sum().reset_index()
```

```python
ax = sns.barplot(x='year', y='total', data=inhabitants_per_year, palette="Blues_d")
ax.set_title('Number of inhabitants per year')
```

```python
columns_to_normalize = ['spanish_men','spanish_women','spanish_total','foreign_men','foreign_women','foreign_total',
                       'men_total','women_total','total']
norm_data = normalize_inhabitants_per_year(inhabitants_per_year, data, columns_to_normalize)
```

#### 1.2. Calculate ratio of variation regarding 2004 and re-scale results to be comparable between segments

```python
def calculate_ratio(data, age_range, ref_year, segment):
    data_ref = data[(data['age_range'] == age_range) & (data['year'] == ref_year)]
    data_target = data[(data['age_range'] == age_range) & ~(data['year'] == ref_year)]
    
    # calculate ratio per district per year
    districts = data_ref['district'].unique().tolist()
    years = data_target['year'].unique().tolist()
    ratios = []
    for year in years:
        ratios_districts = []
        for district in districts:
            ref = data_ref[data_ref['district'] == district][segment].iloc[0]
            target = data_target[(data_target['year'] == year) & (data_target['district'] == district)][segment].iloc[0]
            ratio = target/ref
            ratios_districts.append(ratio)
        ratios.append(ratios_districts)    
    
    # build results data structure
    return pd.DataFrame(ratios, columns=districts, index=years)


def calculate_normalized_cumulative_variation(data):
    # total cumulative
    total_variation = data.sum(axis=0)
    
    # increase and decrease cumulative
    positive_variation = data.apply(lambda x: x[x > 1].sum(), axis=0)
    negative_variation = data.apply(lambda x: x[x < 1].sum(), axis=0)
    
    # normalize dividing by the average variation
    total_mean = total_variation[total_variation != 0].mean()
    positive_mean = positive_variation[positive_variation != 0].mean()
    negative_mean = negative_variation[positive_variation != 0].mean()
    
    total_variation_n = total_variation #/ total_mean
    positive_variation_n = positive_variation #/ positive_mean
    negative_variation_n = negative_variation #/ negative_mean
    
    return total_variation_n, positive_variation_n, negative_variation_n


def scale_matrix(data):
    autoscaler = StandardScaler()
    data_ = data.copy()
    data_ = data_.replace(np.inf, np.nan)
    data_ = data_.fillna(0)
    
    variacion = []
    for i in range(len(data_.index)):
        row = autoscaler.fit_transform(data_.iloc[i].values.reshape(-1, 1))
        variacion.append(list(itertools.chain.from_iterable(row.tolist())))
    variacion = pd.DataFrame(variacion, columns=data_.columns.tolist(), index=data_.index.tolist()) 
    return variacion


def build_heatmap_variations_per_segment_per_age_range(data, age_ranges, segments, ref_year):
    # calculate variation matrices
    variation_matrix_total = {}
    variation_matrix_positive = {}
    variation_matrix_negative = {}
    for segment in segments:
        for age_range in age_ranges:
            column = segment + "_" + age_range
            results = calculate_ratio(data, age_range, ref_year, segment)
            total_variation, positive_variation, negative_variation = calculate_normalized_cumulative_variation(results)
            variation_matrix_total[column] = total_variation
            variation_matrix_positive[column] = positive_variation
            variation_matrix_negative[column] = negative_variation
    
    # build final matrices
    variation_matrix_total = pd.DataFrame.from_dict(variation_matrix_total)
    variation_matrix_positive = pd.DataFrame.from_dict(variation_matrix_positive)
    variation_matrix_negative = pd.DataFrame.from_dict(variation_matrix_negative)
    
    # scale values to same shape
    variation_matrix_total_ = scale_matrix(variation_matrix_total)
    variation_matrix_positive_ = scale_matrix(variation_matrix_positive)
    variation_matrix_negative_ = scale_matrix(variation_matrix_negative)
    
    return variation_matrix_total_, variation_matrix_positive_, variation_matrix_negative_
```

```python
# get parameters
ref_year = 2004
age_ranges = norm_data['age_range'].unique().tolist()[:-1]
segments = norm_data.columns.tolist()[1:10]

# calculate ratio, normalize and rescale
variation_matrix_total, variation_matrix_positive, variation_matrix_negative = build_heatmap_variations_per_segment_per_age_range(norm_data,age_ranges,segments, ref_year)

# order labels by age_range
variation_matrix_total = variation_matrix_total.loc[:,columns_ordered]
variation_matrix_positive = variation_matrix_positive.loc[:,columns_ordered]
variation_matrix_negative = variation_matrix_negative.loc[:,columns_ordered]
```

### 2. Conclusions

* There is a difference in all cases between the patterns seen in foreigners and spaniards.
* The difference between genders is not significant for spaniards, foreigners or all together.
* Paterns observed in the increasing cummulative variation:
    * Increment of the foreign population was seen in a wider age ranges than spaniards
    * Increment of the foreign population was stronger in the range between 30-70 years (active population)
    * Increment of the spanish population was seen in the ranges between 40-60 and between 80-95 years
    * The variation is significantly low for spaniards between 10-30 in all neighborhoods
    * Districts such as Fuencarral, Hortaleza, Usera, Carabanchel or Barajas show a higher increase in the population older than 90 years (lower income neighborhoods)
* Patterns observed in the decreasing cummulative variation:
    * Lower decrease in foreigners than in spaniards
    * High decrease on the population of spaniards in the age range between 10-45 and 60-75 years
    * Significant decrease on the young population in neighborhoods such as Villa de Vallecas, Hortaleza, Fuencarral, Vicalvaro, Barajas and Carabanchel (lower income neighborhoods)
    * Lower decrease on the young spanish population in neighborhoods such as Centro, Moncloa, Tetuan or Salamanca (higher income neighborhoods)
    * Highest decrease on the old spanish population in neighborhoods such as Fuencarral, Hortaleza, Arganzuela, San Blas or Usera (lower income neighborhoods)
    
In general, we see that the active population has increased mainly with foreigners, young people, mainly spanish population, has drastically decrease (likely due to the crisis and therefore, their migration) and the non-active spanish population has increased.

Additionally, besides of the general trends descrived above, we also see differences between the different districts on the spanish population variation. 

The more wealthy districts (Salamanca, Centro, Retiro, Moncloa) are more stable, have less dramatic changes, young people leave less and there is an incrase of the population in a broader age range instead of just old population. 

On the other hand, less wealthy districts (Villa de Vallecas, San Blas or Vicalvaro) show a drastic decrease on the young population and also on population between 60-80, which is interesting (this people is not migrating to other countries to start a new life). This districts, surprisingly, do not show a strong increase in the non-active population as one can expect due to the population pyramid effect.

Finally, middle wealthy districts (Fuencarral, Hortaleza, Chamartin or Barajas) follows the trend of young population decreasing significantly and also population between 60-75. However, there is a strong increase in the population older than 75/80.

We can conclude that there is an effect of the crisis, mainly in the spanish population. Additionally, this effect has a different impact depending on the wealthy level of the district.
